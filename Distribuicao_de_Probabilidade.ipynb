# 2. Conceitos Fundamentais


---



## 2.1 Variável Aleatória


<div align="justify">

Uma variável aleatória é uma função que associa a cada resultado de um experimento aleatório um valor numérico. Essas variáveis podem ser classificadas em dois tipos principais: discretas e contínuas (DEVORE, 2016).

As variáveis aleatórias discretas assumem valores finitos ou enumeráveis, sendo comuns em experimentos que envolvem contagem, como o número de falhas em um conjunto de corpos de prova. Já as variáveis contínuas podem assumir qualquer valor em um intervalo, sendo típicas em medições físicas, como resistência, deformação e deslocamento.

No contexto da engenharia de materiais, a resistência à compressão ou à flexão de um compósito cimentício é um exemplo clássico de variável contínua, enquanto o número de fissuras em um elemento estrutural pode ser tratado como uma variável discreta.


</div>

## 2.2 Distribuição de Probabilidade

<div align="justify">

A distribuição de probabilidade descreve como os valores de uma variável aleatória estão distribuídos, associando probabilidades a cada possível resultado. No caso de variáveis discretas, essa distribuição é definida por uma função de probabilidade; para variáveis contínuas, é descrita por uma função densidade de probabilidade (MONTGOMERY; RUNGER, 2014).

As distribuições de probabilidade permitem não apenas descrever o comportamento dos dados, mas também realizar inferências e previsões, sendo fundamentais no planejamento e análise de experimentos.

</div>

## 2.3 Importância na Engenharia Civil

<div align="justify">

Na engenharia civil, a variabilidade dos materiais e dos processos construtivos torna indispensável o uso de modelos probabilísticos. A análise estatística permite:

- Quantificar a dispersão dos resultados experimentais;
- Avaliar a confiabilidade de materiais e estruturas;
- Comparar diferentes composições ou processos;
- Apoiar a tomada de decisão em projetos e controle tecnológico.

Em materiais cimentícios reforçados com fibras, por exemplo, a variabilidade na distribuição das fibras, na aderência fibra-matriz e nas condições de moldagem influencia diretamente o comportamento mecânico, sendo essencial a utilização de distribuições de probabilidade para uma análise adequada desses sistemas.

#  3. Distribuição Binomial

---



<div align="justify">

A distribuição binomial é uma das principais distribuições de probabilidade discreta, sendo utilizada para modelar experimentos compostos por um número fixo de ensaios independentes, nos quais cada ensaio apresenta apenas dois resultados possíveis: sucesso ou fracasso.

Segundo Walpole *et al.* (2012), a distribuição binomial descreve o número de sucessos em uma sequência de experimentos independentes com probabilidade constante de ocorrência. Essa estrutura é amplamente utilizada em problemas de engenharia que envolvem decisões binárias, como aprovação ou falha de elementos estruturais.

o material apresentado em aula, esse conceito foi ilustrado por meio do lançamento de moedas, onde cada lançamento representa um ensaio e os resultados possíveis são “cara” (sucesso) ou “coroa” (fracasso) . Esse exemplo pode ser diretamente extrapolado para diversas aplicações na engenharia.

</div>


<div style="background-color:#f9f0ff; padding:15px; border-left:5px solid #722ed1">

**Definição:**

Uma variável aleatória \( X \) segue uma distribuição binomial quando representa o número de sucessos em \( n \) ensaios independentes, cada um com probabilidade constante de sucesso \( p \).

</div>



###  Função de Probabilidade

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

<div align="justify">

Onde:

- \( n \): número total de ensaios  
- \( k \): número de sucessos desejados  
- \( p \): probabilidade de sucesso  
- \( (1 - p) \): probabilidade de fracasso  


</div>


<div align="justify">

A distribuição binomial permite avaliar a probabilidade de ocorrência de um determinado número de eventos em um conjunto finito de experimentos. Na engenharia civil, isso é particularmente útil na análise de fenômenos discretos, como a ocorrência de falhas em corpos de prova, presença de fissuras ou conformidade de elementos estruturais com critérios normativos.

</div>

<div style="background-color:#e6f7ff; padding:15px; border-left:5px solid #1890ff">

**Exemplo (Engenharia de Materiais):**

Considere um conjunto de 10 corpos de prova de compósito cimentício reforçado com fibras vegetais. Suponha que a probabilidade de um corpo de prova apresentar fissuração precoce seja de 20% (\( p = 0,2 \)).

Deseja-se determinar a probabilidade de exatamente 3 corpos de prova apresentarem fissuração precoce.

Parâmetros do problema:

- \( n = 10 \)  
- \( k = 3 \)  
- \( p = 0,2 \)  

Esse tipo de abordagem permite avaliar a confiabilidade do material e compreender a variabilidade do comportamento experimental.

</div>

<div style="background-color:#fffbe6; padding:15px; border-left:5px solid #faad14">

**Observação:**

A aplicação da distribuição binomial pressupõe que os ensaios sejam independentes e que a probabilidade de sucesso permaneça constante. Em materiais cimentícios, essas condições podem ser influenciadas por fatores como variabilidade na matriz, distribuição de fibras e condições de moldagem.

</div>

### Implementação Computacional

<div align="justify">

Para complementar a análise teórica, é possível utilizar ferramentas computacionais para calcular probabilidades e visualizar o comportamento da distribuição binomial.

Neste exemplo, será considerada a situação previamente descrita, envolvendo corpos de prova de compósitos cimentícios reforçados com fibras. O objetivo é calcular a probabilidade de ocorrência de diferentes números de falhas (fissuração precoce) e representar graficamente essa distribuição.

A utilização de linguagens de programação, como Python, permite automatizar cálculos e gerar visualizações que facilitam a interpretação dos resultados, sendo uma ferramenta amplamente empregada no planejamento e análise de experimentos.

</div>

```python
# Importação das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parâmetros do problema
n = 10
p = 0.2

# Valores possíveis
k = np.arange(0, n+1)

# Probabilidades
probabilidades = binom.pmf(k, n, p)

# Mostrar valores
for i in range(len(k)):
    print(f"P(X = {k[i]}) = {probabilidades[i]:.4f}")

# Gráfico
plt.figure()
plt.bar(k, probabilidades)
plt.xlabel("Número de falhas")
plt.ylabel("Probabilidade")
plt.title("Distribuição Binomial")
plt.show()
```

### Interpretação dos Resultados

<div align="justify">

O gráfico apresentado ilustra a distribuição de probabilidade do número de corpos de prova que podem apresentar fissuração precoce.

Observa-se que os valores mais prováveis concentram-se em baixos números de falhas, o que é esperado considerando a probabilidade de falha adotada (\( p = 0,2 \)). À medida que o número de falhas aumenta, a probabilidade associada diminui significativamente.

Esse comportamento é coerente com sistemas em que a ocorrência de falhas individuais possui baixa probabilidade, sendo a distribuição binomial uma ferramenta adequada para representar esse tipo de fenômeno.

Na engenharia de materiais, essa análise permite avaliar a confiabilidade de compósitos cimentícios e compreender a variabilidade dos resultados experimentais, auxiliando na tomada de decisões quanto ao desempenho e segurança dos materiais.

##  4. Distribuição Normal

---



<div align="justify">

A distribuição normal é uma das mais importantes distribuições de probabilidade contínua, sendo amplamente utilizada na modelagem de fenômenos naturais e experimentais. Sua relevância decorre do fato de que diversos processos físicos e propriedades de materiais apresentam comportamento aproximadamente normal, especialmente quando influenciados por múltiplos fatores independentes.

Segundo Devore (2016), a distribuição normal é frequentemente empregada para representar variáveis contínuas, como resistência mecânica, deformações e propriedades físicas de materiais, devido à sua capacidade de descrever adequadamente a variabilidade experimental.

No contexto apresentado em aula, observou-se que, à medida que o número de ensaios aumenta, distribuições discretas, como a binomial, tendem a se aproximar da distribuição normal, evidenciando sua importância teórica e prática.

</div>

<div style="background-color:#f9f0ff; padding:15px; border-left:5px solid #722ed1">

**Definição:**

Uma variável aleatória contínua segue uma distribuição normal quando sua função densidade de probabilidade apresenta formato simétrico em torno da média, sendo caracterizada pelos parâmetros média (\( \mu \)) e desvio padrão (\( \sigma \)).

</div>

###  Função Densidade de Probabilidade

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

<div align="justify">

Na expressão da função densidade, a média \( \mu \) define a posição central da distribuição, enquanto o desvio padrão \( \sigma \) controla a dispersão dos valores em torno dessa média. Assim, quanto maior o desvio padrão, mais espalhados estarão os dados; quanto menor, mais concentrados eles estarão próximos ao valor médio.
